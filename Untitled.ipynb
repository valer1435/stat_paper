```python
import pandas as pd
import os
from scipy.stats import *
import scikit_posthocs as sp
import numpy as np
```

```python
def parse_csv_report(path):
    return pd.read_csv(path)[['MAE', 'MAPE', 'Time series label']].values


def get_metric(framework):
    res = {}
    path = os.path.join('sensors_case_1', 'results', framework)
    for name in os.listdir(path):
        if 'report' in name:
            res[int(name.split('_')[0])] = parse_csv_report(os.path.join(path, name))
    return res
```

```python
horizon = 80
fedot_list = list(np.ravel(get_metric('fedot')[horizon][:, 1]))
autots_list = list(np.ravel(get_metric('autots')[horizon][:, 1]))
prophet_list = list(np.ravel(get_metric('prophet')[horizon][:, 1]))
print(np.array(fedot_list).mean())
print(np.array(autots_list).mean())
```

```python
friedmanchisquare(fedot_list, autots_list, prophet_list)
```

По тесту Фрейдмана так как p_value < 0.05 - гипотеза о равном среднем отвергается

```python
sp.posthoc_nemenyi_friedman(np.array([fedot_list, autots_list, prophet_list]).T)
```
